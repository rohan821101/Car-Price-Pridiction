```python
df2 = df1.drop(["doornumber","enginelocation","cylindernumber","drivewheel","fuelsystem","enginetype"],axis=1)
df2
```

```python
x = df2.drop(["price","CarName"],axis=1)
y = df2.pop("price")
```

```python
x
```

```python
x['fueltype']=x['fueltype'].apply(lambda x: 1 if x =='gas' else 0)
x['aspiration']=x['aspiration'].apply(lambda x: 1 if x =='std' else 0)
```

```python
carBody = pd.get_dummies(x.carbody)
CarBrand = pd.get_dummies(x.CarBrand)
```

```python
x = pd.concat([x,carBody,CarBrand],axis=1)
```

```python
x
```

```python
x = x.drop(["carbody","CarBrand"],axis=1)
```

```python
x.info()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()


```

```python
lr_clf.fit(x_train,y_train)
```

```python
lr_clf.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)

cross_val_score(LinearRegression(),x,y,cv=cv)
```
